In der Stadt Luzern gibt es bekanntlich viele Baustellen. Die Dauer X der Arbeiten bei einer Baustelle liege zwischen 0 und 20 Wochen. Die Dichte f(x) habe die folgende Form.


---
a) Begründen Sie, warum c=0.1 ist und schreiben Sie die Dichte f(x) explizit auf.

Das integral der funktion `f(x)` muss 1 geben:

\begin{align}
F(x) = 1 = \int_{0}^{20} f(x) dx
\end{align}

Die Funktion `f(x)` ist von der form:

\begin{align}
f(x) = mx + b
\end{align}

Somit ist `m` und `b`:

\begin{align}
m = \frac{\Delta y}{\Delta x} = \frac{0 - c}{20 - 0} = -\frac{c}{20} \\
b = c \\
f(x) = -\frac{c}{20}x + c \\
\end{align}

Wenn nun `c = 0.1`:

\begin{align}
f(x) -\frac{0.1}{20}x + 0.1\\
0 = f(20) = -\frac{0.1}{20}20 + 0.1
\end{align}

b) Berechnen Sie die Wahrscheinlichkeit, dass die Bauzeit X weniger als **5** && **10** Wochen beträgt.

\begin{align}
P(X \leq \alpha)\rvert_{X = \{5, 10 \}}
\end{align}

Somit nehmen wir die aufleitung von `f(x)`

\begin{align}
F(X) = \int f(x) dx = \int -\frac{0.1}{20}x + 0.1 dx
\end{align}

Jeweils integriert von $0 - 5$ bzw von $0 - 10$:

\begin{align}
\int_{0}^{5} -\frac{0.1}{20}x + 0.1 dx = 0.438 \\ \\
\int_{0}^{10} -\frac{0.1}{20}x + 0.1 dx = 0.75
\end{align}

Das wird in python gerechnet:

```python
from scipy.integrate import quad

f = lambda x: -(0.1/20)*x + 0.1

time_5, time_10 = quad(f, a=0, b=5), quad(f, a=0, b=10)
```

```python
time_5[0], time_10[0]
```

c) Skizzieren Sie die kumulative Verteilungsfunktion

Wir zeichnen also $F(x) = \int f(x) dx$:

\begin{align}
F(x) = \int f(x) dx = \int -\frac{0.1}{20}20 + 0.1 dx = −0.0025x^2+0.1x
\end{align}

In python berechnen wir also:

```python
import sympy

x = sympy.Symbol("x")
sympy.integrate(-(0.1/20)*x + 0.1, x)
```

```python
import numpy as np
from matplotlib import pyplot as plt


x = np.linspace(0, 20, 100)
F = lambda x: -0.0025*x**2 + 0.1*x

plt.plot(x, F(x))
```
