Monte Carlo Algorithmen sind randomisierte Algorithmen und stellen ein gutes Werkzeug für Simulationen von stochastischen Prozessen dar.  Auch die Zahl π lässt sich mit Hilfe von Monte  Carlo Simulationen  bestimmen. Im Folgenden möchten wir ein Computerprogramm erstellen,  mit welchem man die Zahl π aufgrund von MonteCarlo Methoden simulieren kann. Man generiert hierzu zufällige Punkte

und überprüft, ob diese innerhalb des Einheitskreises  mit Kreismittelpunkt MK=(0, 0)und Radius r=1 liegen. Die sich ergebende Wahrscheinlichkeitsverteilung P\[(x,y)∈Kreis\] stellt die Fläche eines Viertels des Einheitskreises dar. π kann nun mit folgender Formel berechnet werden



Bestimmen Sie mit Hilfe dieser Überlegung die Zahl π. Python-Hinweise:

```python
# Generieren von 100 gleichmaessig verteilten
# Zufallszahlen im Intervall [-1,1]
from scipy.stats import uniform
import numpy as np

x=uniform.rvs(size=100, loc=-1, scale=2)
# Bestimmen der Anzahl Zahlen kleiner als eins; Beispiel:
# Anzahl von 100 zufaellig im Intervall [0,10] generierten
# Zahlen, die kleiner als 1 sind:
np.sum(x<1)
```

```python
total = 10000000

x = uniform.rvs(size=total, loc=-1, scale=2)
y = uniform.rvs(size=total, loc=-1, scale=2)

absolute = np.sum(x**2 + y**2 <= 1)

absolute/total * 4
```
