**teil b**

Bestimmen Sie die Parameter a und b des linearen Modells `y = a + bx`, wobei y
das Einkommen bezeichnet und x die Anzahl Jahre Schulbildung bzw. AFQT.
Zeichnen Sie die Regressionsgerade wie im Skript:

```
x = np.linspace(... , ...)
plt.plot(x, a+b*x, c="orange")
```

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.linspace(income["Income2005"], income["AFQT"])
````

````python
a, b = np.polyfit(income["Income2005"], income["AFQT"], deg=1)

plt.plot(x, a+b*x, c="orange")
plt.show()
````

````python
a, b = np.polyfit(income["Income2005"], income["Educ"], deg=1)

plt.plot(x, a+b*x, c="orange")
plt.show()
````

Wie interpretieren Sie jeweils die Parameter `a` und `b`?

`a` bezeichnet die x abweichung

`b` bezeichnet die y abweichung
